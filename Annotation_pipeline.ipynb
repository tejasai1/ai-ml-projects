```python
import json
```

```python
dataset = [
    {"id": 1, "text": "The app crashed on startup", "label": "bug"},
    {"id": 2, "text": "ok",                          "label": "feedback"},
    {"id": 3, "text": "Add dark mode please",         "label": "feature"},
    {"id": 4, "text": "How do I export data?",        "label": "question"},
    {"id": 5, "text": "Login button is broken",       "label": "bug"},
]
```

```python
label_counts = {}
for item in dataset:
    label = item["label"]
    label_counts[label] = label_counts.get(label, 0) + 1
print("Label counts:", label_counts)
```

```python
cleaned = [item for item in dataset if len(item["text"]) >= 5]
print(f"Kept {len(cleaned)} of {len(dataset)} items after filtering")
```

```python
with open("output.json", "w") as f:
    json.dump(cleaned, f, indent=2)
print("Saved to output.json")
```

```python
with open("output.json", "r") as f:
    result = json.load(f)
print("\nFinal output:")
for item in result:
    print(item)
```
